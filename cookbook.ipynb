## First recipe: utilize the functions
### It is good thing to use in python is functions; they make the life easier for you and you can reuse them whenver you need.

#### In the following example, I wanted to devide the data set into 10 groups according to the hidden gem score. The range of scores is from zero to ten.

```python
# First, import the data set
import pandas as pd
netflix = pd.read_csv('netflix-rotten-tomatoes-metacritic-imdb.csv')
netflix
```

```python
# Now, create the function
def score(x):
    if x <= 1:
        return '0-1 score'
    elif x <=2:
        return '1-2 score'
    elif x <=3:
        return '2-3 score'
    elif x <=4:
        return '3-4 score'
    elif x <=5:
        return '4-5 score'
    elif x <=6:
        return '5-6 score'
    elif x <=7:
        return '6-7 score'
    elif x <=8:
        return '7-8 score'
    elif x <=9:
        return '8-9 score'
    elif x <=10:
        return '9-10 score'
    else:
        return "Unrated"
    
netflix['gem score range'] = netflix['Hidden Gem Score'].apply(score)
netflix
```

### There is another way to do it, but things are much easier using the functions.

```python
#Another way to do it
ranges = []
for x in netflix['Hidden Gem Score']:
    if x <= 1:
        ranges.append('0-1 score')  
    elif x <=2:
        ranges.append('1-2 score')
    elif x <=3:
        ranges.append('2-3 score')
    elif x <=4:
        ranges.append('3-4 score')
    elif x <=5:
        ranges.append('4-5 score')
    elif x <=6:
        ranges.append('5-6 score')
    elif x <=7:
        ranges.append('6-7 score')
    elif x <=8:
        ranges.append('7-8 score')
    elif x <=9:
        ranges.append('8-9 score')
    elif x <=10:
        ranges.append('9-10 score')
    else:
        ranges.append("Unrated")
#add the list to the data set
netflix['gem score range 2'] = ranges
netflix
```

#### We see from the results that both ways worked. However, the function approach is cleaner and reusable if want to apply for many places in our code.

## Second recipe: Use isin() function:
### The isin() function is so good to return a boolean value when checking if something exists in a certain column in your data set.

#### In the following example, I wanted to focus on directors with more than 10 shows.
#### I first created a list Directors_new that has the directors with more than ten shows.

```python
Directors = netflix['Director'].value_counts()
Directors_new = []
for x in Directors.index:
    if Directors[x] >= 10:
        Directors_new.append(x)
Directors_new
```

```python
# Now, apply the isin()
netflix_dr = netflix[netflix['Director'].isin(Directors_new)]

# This will create a new data set that only has shows directed by the directors in our list.
netflix_dr
```

#### So, this is a good way when you want to check a condition for a whole column in your data set.

## Third recipe: take the advantage of date-time data type.
### Data-time data is so easy to deal with if you are working on time related columns.


```python
df = pd.read_csv("time_stamp.csv")
df
```

#### I want to create a column that has the time difference between each row. It's very hard to that by ourselves, so we convert our values to datetime to benefit from the special functions that datetime has.

```python
df["endTime"] = pd.to_datetime(df["endTime"])
df
```

```python
df["Diff"] = df["endTime"].diff()
df
```

#### And there are more feature associated with datetime values.
